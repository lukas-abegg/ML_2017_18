## Task b) Simple Linear Regression

```python
import scipy.stats as pstats
def lin_reg(x, y):
    r2 = 0
    xm = np.mean(x)
    ym = np.mean(y)
    #compute b_1
    fst = 0
    snd = 0
    for i,v in enumerate(x):
        fst += (x[i]-xm)*(y[i]-ym)
        snd += (x[i]-xm)**2
    b_1 = fst/snd
    #compute b_0
    b_0 = ym - b_1 * xm
    #compute y_r
    y_r = []
    for i,v in enumerate(x):
        y_r.append(b_0 + b_1 * x[i])
    #compute RÂ² and SE of Slope
    fst = 0
    snd = 0
    thd = 0
    for i,v in enumerate(y):
        fst += (y[i] - y_r[i])**2
        #important to compute r
        snd += (y[i] - ym)**2
        #important to compute SE
        thd += (x[i] - xm)**2
    r2 = 1 - fst/snd
    SE = np.sqrt(fst/(len(y)-2)) / np.sqrt(thd)
    t = np.abs(b_1/SE)
    pval = pstats.t.sf(np.abs(b_1/SE), len(y)-2)
    #Pearson correlation and p value for non-correlation
    _, rpval = pstats.pearsonr(x,y)
    #quadric loss
    loss = 0
    for i,v in enumerate(y):
        loss += (y[i] - y_r[i])**2
    return b_1, b_0, np.sqrt(r2), t, pval, rpval, loss, y_r
```

```python
#Compute all possible simple linear regression models
for i in columns[:-1]:
    b1, b0, r2, t, pval, rpval, loss, _ = lin_reg(list(boston[i]), list(boston['MEDV']))
    print ""
    print "---",i,"---"
    print "Slope:",b1, "\nIntercept:",b0
    print "Coefficient of Determination:",r2
    print "t-statistic of slope:", t
    print "p-Value:", pval
    print "pearson p-Value:", rpval
    print "Quadric Loss:",loss
```

According to R-Measure the linear model with features RM, PTRATIO, LSTAT explains most of the variation of data.
According to p-Value the nearest models to significant fit are also RM, PTRATIO, LSTAT.
Therefor we decide that those features fit best to combine for a multiple linear regression.

```python
#plotting function for simple linear regression
def plotreg(x,y):
    _, _, _, _, _, _, _, y_r = lin_reg(x,y)
    plt.plot(x,y_r)
    plt.scatter(x,y)
    plt.show()
    plt.close()
```

```python
plotreg(list(boston['RM']), list(boston['MEDV']))
plotreg(list(boston['PTRATIO']), list(boston['MEDV']))
plotreg(list(boston['LSTAT']), list(boston['MEDV']))
```

## Task c) Multiple Linear Regression

```python
def multi_lin_regr(X, Y):
    #Preprocess to convert for Matrix-Operations
    X = np.stack(X, axis=1)
    X = np.matrix(X)
    Y = np.matrix(Y).transpose()
    #compute coefficent-vector b
    b = np.linalg.inv(X.transpose()*X)*X.transpose()*Y
    #compute modeled Y
    Y_r = X*b
    #compute quadric loss
    #dist = (Y-Y_r)
    loss = 0
    dist = []
    for i,v in enumerate(Y.tolist()[0]):
        dist.append((Y.item(i)-Y_r.item(i))**2)
    for i,v in enumerate(dist):
        loss += dist[i]
    return b, loss
```

```python
#Take the three significantly linear modelled features (according to p-value) and compute multiple regression models
Y = np.array(list(boston['MEDV']))
x1 = np.array(list(boston['RM']))
x2 = np.array(list(boston['PTRATIO']))
x3 = np.array(list(boston['LSTAT']))
X = np.stack((x1,x2))
print "linear regression with features RM, PTRATIO\n", multi_lin_regr(X, Y)
X = np.stack((x1,x3))
print "linear regression with features RM, LSTAT\n", multi_lin_regr(X, Y)
X = np.stack((x2,x3))
print "linear regression with features PTRATIO, LSTAT\n", multi_lin_regr(X, Y)
```

The combination of the features with smallest p-values has the smallest loss.
