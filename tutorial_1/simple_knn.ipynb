```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from sklearn import utils
from sklearn.model_selection import train_test_split

# Generate two-dimensional training data and test data coming from two classes 
count_per_class = 50
df_A = pd.DataFrame(np.random.randint(0,60,size=(count_per_class, 2)), columns=list('xy'))
df_A['class'] = pd.Series("A", index=df_A.index)

df_B = pd.DataFrame(np.random.randint(40,100,size=(count_per_class, 2)), columns=list('xy'))
df_B['class'] = pd.Series("B", index=df_B.index)

df = pd.concat([df_A, df_B])

df_data = df.loc[:, ['x', 'y']].values
df_label = df.loc[:, ['class']].values

# Split test and training data
data_train, data_test, label_train, label_test = train_test_split(df_data, df_label, test_size=0.33, random_state=42)

print 'Amount of training data: ' + str(len(data_train))
print 'Amount of test data: ' + str(len(data_test))
```

```python
# calculate euclidean distance
def calculate_distance(elem1, elem2):
    distance = 0
    # use x and y coordinates of both points to calculate distance
    for x in range(len(elem1)):
        distance += pow((elem1[x] - elem2[x]), 2)
    return math.sqrt(distance)

# sort by distances 
def sort_value(elem):
    return elem[1]

# get top k nearest neighbors
def get_k_neighbors(data_train, label_train, elem, k):
    dists = []
    # calculate all distances from point elem to all others
    for x in range(len(data_train)):
        # use structure (label, euclidean distance value) to save distances from all other points
        dists.append((label_train[x], calculate_distance(elem, data_train[x])))
    # sort by distance value ascending
    dists.sort(key=sort_value)
    # return k nearest neighbors
    return dists[:k]

# predict label by knn classifier
def classify(neighbors, elem):
    # count nearest neighbors from both classes to classify point elem
    a = len(filter(lambda x: x[0] == 'A', neighbors))
    b = len(filter(lambda x: x[0] == 'B', neighbors))
    # classify based on occurrences of both classes in nearest neighbors
    if a > b:
        return 'A'
    else:
        return 'B'

# predict labels for all points from test_data and return them
def predict_data(data_test, k):
    predicted = []
    for x in range(len(data_test)):
        elem = data_test[x]
        # use structure (x, y, label) to save predictions
        predicted.append((elem[0], elem[1], classify(get_k_neighbors(data_train, label_train, elem, k), elem)))
    return predicted

# predict all test points for k=5
predicted = predict_data(data_test, 5)
```

```python
# prepare data for plots
train_data_labeled = np.concatenate((data_train, label_train), axis=1)

# group train data by class
train_A = np.array(filter(lambda x: x[2] == 'A', train_data_labeled))
train_B = np.array(filter(lambda x: x[2] == 'B', train_data_labeled))

df_trainA = pd.DataFrame(train_A)
df_trainB = pd.DataFrame(train_B)

# group test data by class
predicted_A = np.array(filter(lambda x: x[2] == 'A', predicted))
predicted_B = np.array(filter(lambda x: x[2] == 'B', predicted))

df_predA = pd.DataFrame(predicted_A)
df_predB = pd.DataFrame(predicted_B)
```

```python
# plot result as 2D figure
plotA = plt.plot(pd.to_numeric(df_trainA[0]), pd.to_numeric(df_trainA[1]), "ro", label="class A")
plotB = plt.plot(pd.to_numeric(df_trainB[0]), pd.to_numeric(df_trainB[1]), "bo", label="class B")

plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(-10, 110)
plt.ylim(-10, 110)
plt.legend(bbox_to_anchor=(1, 1))
plt.title("Plot for test data")
plt.show()
```

```python
# Build KNN decision boundary
#------------------------------

def calculate_boundary_for_k(k):
    # build test data matrix
    boundary = []
    for x in range(101):
        for y in range(101):
            boundary.append([x,y])

    # classify each point of matrix
    boundary_predicted = predict_data(boundary, k)

    # prepare data for plots
    boundary_predicted_A = np.array(filter(lambda x: x[2] == 'A', boundary_predicted))
    boundary_predicted_B = np.array(filter(lambda x: x[2] == 'B', boundary_predicted))

    df_boundary_predicted_A = pd.DataFrame(boundary_predicted_A)
    df_boundary_predicted_B = pd.DataFrame(boundary_predicted_B)

    # plot result as 2D figure
    plotA_pred = plt.plot(pd.to_numeric(df_boundary_predicted_A[0]), pd.to_numeric(df_boundary_predicted_A[1]), "rx",alpha=.25, label="class A boundary")
    plotB_pred = plt.plot(pd.to_numeric(df_boundary_predicted_B[0]), pd.to_numeric(df_boundary_predicted_B[1]), "bx",alpha=.25, label="class B boundary")
    # override train data points with thick points to visualize them
    plotA = plt.plot(pd.to_numeric(df_trainA[0]), pd.to_numeric(df_trainA[1]), "ro", label="class A")
    plotB = plt.plot(pd.to_numeric(df_trainB[0]), pd.to_numeric(df_trainB[1]), "bo", label="class B")

    plt.xlabel('X')
    plt.ylabel('Y')
    plt.xlim(-10, 110)
    plt.ylim(-10, 110)
    plt.legend(bbox_to_anchor=(1, 1))
    plt.title("Plot decision boundary for k=" + str(k))
    plt.show()
```
