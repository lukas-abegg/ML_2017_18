```python
import numpy as np
import pandas as pd
import csv
from random import shuffle
```

```python
table = []
f = open('spambase.data')
for i in f.readlines():
    table.append(np.float64((i.split('\n'))[0].split(',')))
```

```python
columns = []
f = open('spambase.names')
for i in f.readlines()[33:]:
    columns.append(i.split(': ')[0])
```

```python
columns = columns + ['spam']
```

```python
spamdata = pd.DataFrame(table, columns=columns)
spamdata
```

```python
temp = spamdata['spam']
#spamdata = spamdata/100
spamdata['capital_run_length_average'] = spamdata['capital_run_length_average']/100
spamdata['capital_run_length_longest'] = spamdata['capital_run_length_longest']/100
spamdata['capital_run_length_total'] = spamdata['capital_run_length_total']/100

#Norm every column (as mentioned in solution by Patrick)
def normalizeC(d):
    #mu = mean(data)
    #(data .- mu) ./ stdm(data, mu)
    #m = np.mean(d)
    #(d-m) / np.std(d,m)
    mini = np.min(d)
    maxi = np.max(d)
    for i,x in enumerate(d):
        d[i] = d[i] - mini / maxi - mini
    return d

for i in spamdata.columns:
    spamdata[i] = normalizeC(spamdata[i])

spamdata['spam'] = temp
spamdata
```

```python
#get train and test dataset
indices = list(range(0, len(spamdata.index)))
shuffle(indices)
traind = pd.DataFrame(data=np.zeros((0, len(columns))), columns=columns)
testd = pd.DataFrame(data=np.zeros((0, len(columns))), columns=columns)
for i in range(0,len(spamdata.index)):
    if i%2==0:
        traind.ix[i] = spamdata.ix[indices[i]]
    else:
        testd.ix[i] = spamdata.ix[indices[i]]
```

```python
def computeGrad(X, Y, B):
    Bgrad = np.matrix(np.zeros(len(traind.columns))).transpose()
    #compute the gradient over all target-scalars and x-Vectors
    for i, yi in enumerate(Y):
        hBxi = np.exp(B.transpose()*X[:,i]) / (1 + np.exp(B.transpose()*X[:,i]))
        Bgrad = Bgrad + (X[:,i]*(yi - hBxi))
    return Bgrad

#likelihood(theta, X, Y) = sum([Y[i]*dot(theta,X[i,:]) - log(1 + exp(dot(theta,X[i,:]))) for i in 1:size(X,1)])
def llikelihood(X, Y, B):
    ll = 0 #np.matrix(np.zeros(len(traind.columns))).transpose()
    for i,v in enumerate(Y):
        #print(ll)
        ll += ( Y[i] * (B.transpose()*X[:,i]) - np.log( 1 + np.exp(B.transpose()*X[:,i]) ) )
    return ll

def fit(X, Y, B, step):
    loopend = 100
    oldLL = -1e100
    #compute stepwise the maximized coefficient-Vector
    while 1:
        #maybe not this computation but the log likelihood should be B and Bold:
        #B = Bold + slopestep * computeGrad(X, Y, Bold)
        B += step * computeGrad(X, Y, B)
        LL = llikelihood(X, Y, B) #Y*grad.transpose()*X - np.log(1 + np.exp(grad.transpose()*X))
        #print((oldLL-LL)/oldLL)
        loopend -= 1
        if ((oldLL-LL)/oldLL) < 0.01: #or loopend==0:
            break
        oldLL = LL
    return B

def predict(Xt, Yt, B):
    #initilize predict-vector
    pYX = []
    #loop through X-Matrix, for each xi compute probability (pyx) and multiply it to predict-vector
    for i, yi in enumerate(Yt):
        hBxi = np.exp(B.transpose()*Xt[:,i]) / (1 + np.exp(B.transpose()*Xt[:,i]))
        pYX.append(np.asscalar(hBxi))
    #return predict-vector, computed over all xi
    return pYX

def logregress(X, Y, Xt, Yt, Binit, step):
    #insert padding
    padding = np.ones((len(Y)))
    paddingt = np.ones(len(Yt))
    X = np.vstack([padding, np.matrix((X).as_matrix()).transpose()])
    Xt = np.vstack([paddingt, np.matrix((Xt).as_matrix()).transpose()])
    Y = np.matrix(Y.as_matrix()).transpose()
    Yt = np.matrix(Yt.as_matrix()).transpose()
    #fit and predict
    B = fit(X, Y, Binit, step)
    pred = predict(Xt, Yt, B)
    return pred, Yt, B
```

```python
pred, Yreal, teta = logregress(traind.drop('spam', 1),
                               traind['spam'],
                               testd.drop('spam', 1),
                               testd['spam'],
                               np.matrix(np.ones(len(traind.columns))).transpose()*3,
                               0.0001)
print(pred, Yreal, teta)
```

```python
#F-Measure
def fmeasure(Yp, Yr, b):
    TP = 0
    FP = 0
    TN = 0
    FN = 0
    for i in range(0,len(Yp)):
        #print(Yr[i], Yp[i])
        if Yr[i]==1 and Yp[i]==1:
            TP += 1
        elif Yr[i]==0 and Yp[i]==1:
            FP += 1
        elif Yr[i]==0 and Yp[i]==0:
            TN += 1
        elif Yr[i]==1 and Yp[i]==0:
            FN += 1
    print(TP,FP,TN,FN)
    precision = TP/(TP+FP)
    recall = TP/(TP+FN)
    Fb = (1+b**2) * (precision*recall) / ((b**2)*precision+recall)
    return Fb
```

```python
print(fmeasure(np.round(pred),Yreal,0.5))
print(fmeasure(np.round(pred),Yreal,1))
```

```python
def fitRidge(X, Y, B, step, lmbda):
    loopend = 100
    oldLL = -1e100
    #compute stepwise the maximized coefficient-Vector
    while 1:
        B += step * computeGrad(X, Y, B)
        LL = llikelihood(X, Y, B) - (lmbda**2) * B.transpose()*B
        print(loopend, (oldLL-LL)/oldLL, LL)
        loopend -= 1
        if ((oldLL-LL)/oldLL) < 0.01 or loopend==0:
            break
        oldLL = LL
    return B 

def logregressR(X, Y, Xt, Yt, Binit, step, lmbda):
    #insert padding
    padding = np.ones((len(Y)))
    paddingt = np.ones(len(Yt))
    X = np.vstack([padding, np.matrix((X).as_matrix()).transpose()])
    Xt = np.vstack([paddingt, np.matrix((Xt).as_matrix()).transpose()])
    Y = np.matrix(Y.as_matrix()).transpose()
    Yt = np.matrix(Yt.as_matrix()).transpose()
    #fit and predict
    B = fitRidge(X, Y, Binit, step, lmbda)
    pred = predict(Xt, Yt, B)
    return pred, Yt, B
```

```python
predR, YrealR, tetaR = logregressR(traind.drop('spam', 1),
                               traind['spam'],
                               testd.drop('spam', 1),
                               testd['spam'],
                               np.matrix(np.ones(len(traind.columns))).transpose()*3,
                               0.0001,
                               0.2)
print(predR, YrealR, tetaR)
```
