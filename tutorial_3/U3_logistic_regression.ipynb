```python
def manhattan(a,b):
    d = 0
    for i,v in enumerate(a):
        d += np.abs(a[i]-b[i])
    return d

def knn(in_fit, color, in_t, k):
    df = in_fit.copy(deep=True)
    df_t = in_t.copy(deep=True)
    cl_predict = []
    colors = df["income"].map(color)
    for i in range(0,len(df_t["age"])):
        #get testdata
        t1 = list(df_t.iloc[i])
        #compute all distances to datapoints
        distances = []
        for i in range(0,len(df["age"])):
            distances.append(manhattan(df.iloc[i][0:11], t1))
        #get k lowest distances, save kth lowest as threshold
        thresh = sorted(distances)[:k][k-1]
        #get labels of the k datapoints with lowest distances 
        df['Distance'] = distances
        l_list = df[df['Distance']<=thresh][:k]['income']
        #count labels
        counts = {}
        for i in list(set(df['income'])):
            counts[i] = list(l_list).count(i)
        #classify
        cl_predict.append(max(counts, key=counts.get))
        df.drop('Distance', axis=1)
    #plot feature-space with test-data
    df_t["predict"] = cl_predict
    return df_t
```

```python
def computeGrad(X, Y, B):
    Bgrad = np.matrix(np.zeros(len(traind.columns))).transpose()
    #compute the gradient over all target-scalars and x-Vectors
    for i, yi in enumerate(Y):
        hBxi = np.exp(B.transpose()*X[:,i]) / (1 + np.exp(B.transpose()*X[:,i]))
        Bgrad = Bgrad + (X[:,i]*(yi - hBxi))
    return Bgrad

def fit(X, Y, Binit, step):
    slopestep = step
    loopend = 10
    B = Binit
    #compute stepwise the maximized coefficient-Vector
    while 1:
        Bold = B
        B = Bold + slopestep * computeGrad(X, Y, Bold)
        loopend -= 1
        if (np.abs(Bold-B)).any() < 0.0000001 or loopend==0:
            break
    return B
```

```python
def predict(Xt, Yt, B):
    #initilize predict-vector
    pYX = []
    #loop through X-Matrix, for each xi compute probability (pyx) and multiply it to predict-vector
    for i, yi in enumerate(Yt):
        hBxi = np.exp(B.transpose()*Xt[:,i]) / (1 + np.exp(B.transpose()*Xt[:,i]))
        pYX.append(hBxi)
    #return predict-vector, computed over all xi
    return pYX
```

```python
def logregress(X, Y, Xt, Yt, Binit, step):
    B = fit(X, Y, Binit, step)
    pred = predict(Xt, Yt, B)
    return pred, Yt, B
```

```python
padding = np.ones((len(traind['income'])))
paddingt = np.ones(len(testd['income']))
```

```python
pred, Yreal, teta = logregress(np.vstack([padding, np.matrix((traind.drop('income', 1)).as_matrix()).transpose()]), 
                np.matrix(traind['income'].as_matrix()).transpose(),
                np.vstack([paddingt, np.matrix((testd.drop('income', 1)).as_matrix()).transpose()]),
                np.matrix(testd['income'].as_matrix()).transpose(),
                np.matrix(np.ones(len(traind.columns))).transpose(), 
                0.0001)
```
